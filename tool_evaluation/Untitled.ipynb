```python
import os
import glob
import shutil
import matplotlib.pyplot as plt
%matplotlib inline
from extract_crops import save_crops
import pyrealsense2 as rs
import numpy as np
import cv2
import os
import argparse
total=0
def get_pipeline(directory,res=720):
    config = rs.config()
    
    rs.config.enable_device_from_file(config, directory, False)
    pipeline = rs.pipeline()
    
    if res==720: 
        config.enable_stream(rs.stream.depth, 1280,720, rs.format.z16, 30)

        config.enable_stream(rs.stream.color, 1280, 720, rs.format.rgb8, 30)
    else: 
        config.enable_stream(rs.stream.depth, 640,480, rs.format.z16, 30)

        config.enable_stream(rs.stream.color, 848, 480, rs.format.rgb8, 30)
    align_to = rs.stream.color
    align = rs.align(align_to)
    return pipeline,config,align

def process_bag(path,classs):
    first_ioi=None
    first_found=True
    global total
    LOAD_BAG = True
    SAVE_DEPTH = True
    SAVE_RGB = True
    save_dir="unet_crops/"
    listf = []  
    seven=False 
    try: 
        pipeline,config,align=get_pipeline(path,res=480)
        pipeline.start(config)
        seven = True
    except Exception as e:
        print (e)
        
        return
    
    ii = 0
    while True: 
        try:
            frames = pipeline.wait_for_frames()
            aligned_frames = align.process(frames)
            
        except:
            return
        #listf.append(frames)
        ii += 1   
        if SAVE_DEPTH:
            total+=1 
            try:
                depth_frame = aligned_frames.get_depth_frame()
     
            except:
                continue
            depth_image = np.asanyarray(depth_frame.get_data()) 
            
        if SAVE_RGB: 
             
            color_frame = aligned_frames.get_color_frame()
            color_image = np.asanyarray(color_frame.get_data())
        box,_ = (save_crops(color_image,depth_image,"unet"))
        
        if box is not None:
            
            dirs=os.path.join(save_dir,classs)
            if not os.path.exists(dirs):
                os.mkdir(dirs)
            
            file=os.path.join(os.path.join(save_dir,classs),(str(ii)+".jpg"))
           
            if first_found:
                print (path.replace("bag","txt"),ii)
                #np.savetxt(path.replace("bag","txt"),np.array([ii,box[0],box[1],box[2],box[3]]).reshape(-1,5))
                first_found=False
                #return
            cv2.imwrite(file,cv2.cvtColor(color_image[box[1]:box[3],box[0]:box[2]],cv2.COLOR_BGR2RGB))
            
            #cv2.rectangle(color_image,(box[0],box[1]),(box[2],box[3]),(0,255,0),5)
        
        #plt.imshow(color_image.astype("uint8"))
        #plt.show()
        
         
root_dir="./data/"
i=0
folders=os.listdir(root_dir)
for folder in folders:
    classs=folder.split("-")[0]
    folder=os.path.join(root_dir,folder)
    files= (os.listdir(folder))
    
    bag_found=False
    for file in files:
        file=os.path.join(folder,file)
        file_size= (os.path.getsize(file))
        if file_size > 515969418 and "bag" in file:
            i+=1
            process_bag(file,classs) 
```
